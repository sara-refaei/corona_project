```python
#import libraries
import requests
import csv
```

```python
response = requests.get('https://coronavirus.m.pipedream.net/')
```

```python
corona_data = response.json()
```

```python
corona_data['rawData']
```

```python
#create csv file 
with open('corona data.csv', 'w', newline ='') as f:
    writer = csv.DictWriter(f, fieldnames = ['country name', 'province', 'confirmed', 'death'])
    writer.writeheader()
    
    for c in corona_data['rawData']:
        country_name = c['Country_Region']
        province = c['Province_State']
        confirmed = c['Confirmed']
        death = c['Deaths']
        
        writer.writerow(
        {'country name': country_name,
        'province': province,
        'confirmed' : confirmed,
        'death' : death}
        )
```
